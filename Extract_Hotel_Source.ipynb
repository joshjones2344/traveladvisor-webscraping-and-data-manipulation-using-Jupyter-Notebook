```python
import pandas as pd

# Load the CSV file into a DataFrame
df = pd.read_csv(r'C:\Users\Admin\Downloads\SPARTA\SP702 CAPSTONE\datasets\dataset_tripadvisor_2023-hotel.csv')
df
```

 we need to concatenate all amenities into one column only

```python
# Specify the columns you want to concatenate
columns_to_concat = [f'amenities/{i}' for i in range(0, 132)]

# Concatenate the specified columns with a separator, converting non-string elements to strings
separator = ' , '
df['amenities'] = df[columns_to_concat].apply(lambda x: separator.join(str(item) for item in x), axis=1)

# Drop the original columns if needed
df = df.drop(columns=columns_to_concat)

```

```python
df
```

```python
#we need to replace missing values with specific global constant 
df['website'] = df['website'].fillna('NoWebsiteAvailable')
df['email'] = df['email'].fillna('NoEmailAvailable')
last_column_name = 'amenities'
df_columns = df.columns.tolist()
df_columns.remove(last_column_name)

# Reorder the columns with 'amenities' in the 3rd position
new_column_order = df_columns[:2] + [last_column_name] + df_columns[2:]
df = df[new_column_order]
```

```python
# Save the result to a Hotel_DataSet_cleaned.csv
df.to_csv(r'C:\Users\Admin\Downloads\SPARTA\SP702 CAPSTONE\datasets\Hotel_Dataset_Cleaned.csv', index=False)
```
