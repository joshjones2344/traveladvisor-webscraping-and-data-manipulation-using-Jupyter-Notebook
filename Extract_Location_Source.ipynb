```python
import pandas as pd
from fuzzywuzzy import fuzz
from fuzzywuzzy import process

# Read your CSV files
df1 = pd.read_csv(r'C:\Users\Admin\Downloads\SPARTA\SP702 CAPSTONE\FINAL_OUTPUT\Location.csv')
df2 = pd.read_csv(r'C:\Users\Admin\Downloads\SPARTA\SP702 CAPSTONE\datasets\Population_cebu_per_municipality.csv')

# Function to find the best match for each 'address' in df1 in df2
def find_best_match(row, choices):
    return process.extractOne(row['address'], choices, scorer=fuzz.partial_ratio)

# Create a mapping of addresses from df1 to df2
df1['address_mapping'] = df1.apply(find_best_match, args=(df2['address'],), axis=1)

# Create a DataFrame with the mapping results
address_mapping_df = df1['address_mapping'].apply(pd.Series)
address_mapping_df.columns = ['best_match', 'score', 'index']

# Merge df1 with df2 based on the best_match column
merged_df = pd.merge(df1, address_mapping_df, left_index=True, right_on='index')
merged_df = pd.merge(merged_df, df2, left_on='best_match', right_on='address', how='inner')

# Drop the temporary 'address_mapping' column
merged_df = merged_df.drop('address_mapping', axis=1)

# Now merged_df contains the merged data based on the best fuzzy matches of 'address'
```
