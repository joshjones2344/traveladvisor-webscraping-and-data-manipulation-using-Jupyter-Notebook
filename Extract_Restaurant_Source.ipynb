```python
import pandas as pd

# Load the CSV file into a DataFrame
restau_data = pd.read_csv(r'C:\Users\Admin\Downloads\SPARTA\SP702 CAPSTONE\datasets\dataset_tripadvisor_Restaurant.csv')
restau_data
```

we need to concatenate all Cuisine into one column only

```python
# Specify the columns you want to concatenate
columns_to_concat = [f'cuisines/{i}' for i in range(0, 7)]

# Concatenate the specified columns with a separator, converting non-string elements to strings
separator = ' , '
restau_data['Specialties'] = restau_data[columns_to_concat].apply(lambda x: separator.join(str(item) for item in x), axis=1)

# Drop the original columns if needed
restau_data = restau_data.drop(columns=columns_to_concat)
```

```python
restau_data
```

```python
#we need to replace missing values with specific global constant 
restau_data['priceRange'] = restau_data['priceRange'].fillna('NotAvailable')
restau_data['website'] = restau_data['website'].fillna('NoWebsiteAvailable')
last_column_name = 'Specialties'
restau_data_columns = restau_data.columns.tolist()
restau_data_columns.remove(last_column_name)

# Reorder the columns with 'amenities' in the 3rd position
new_column_order = restau_data_columns[:2] + [last_column_name] + restau_data_columns[2:]
restau_data = restau_data[new_column_order]
```

```python
# Save the result to a Restaurant_DataSet_cleaned.csv
restau_data.to_csv(r'C:\Users\Admin\Downloads\SPARTA\SP702 CAPSTONE\datasets\Tourist_Dataset_Cleaned.csv', index=False)
```
