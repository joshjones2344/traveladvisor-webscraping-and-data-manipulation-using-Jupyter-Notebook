```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Define the URL of the web page you want to scrape
url = "https://www.philatlas.com/visayas/r07/cebu.html"
response = requests.get(url)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    html_content = response.text
else:
    print("Failed to retrieve the webpage.")
```

```python
soup = BeautifulSoup(html_content, 'html.parser')

# Find the table element by its HTML tag (e.g., 'table') and any relevant attributes
table = soup.find('table', {'class': 'sortable datatable'})
```

```python
df = pd.read_html(str(table))
```

```python
desired_df = df[0]
```

```python
# Extract columns 0, 1, and 2
desired_df = desired_df.iloc[:, :3]

# Remove the last row (row 50)
desired_df = desired_df.iloc[:-1]

# Now, desired_df contains the desired data
```
