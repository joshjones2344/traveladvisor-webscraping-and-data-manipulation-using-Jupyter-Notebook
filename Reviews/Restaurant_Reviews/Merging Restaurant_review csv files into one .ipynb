```python
import pandas as pd
import glob
from tqdm import tqdm
# Create an empty DataFrame to store the combined data
combined_df = pd.DataFrame(columns=['restaurantname', 'Reviews'])

# List all CSV files in the current directory
csv_files = glob.glob('*.csv')

# Loop through each CSV file and concatenate them
for csv_file in csv_files:
    df = pd.read_csv(csv_file)
    combined_df = pd.concat([combined_df, df])
tqdm.pandas()
# Reset the index of the combined DataFrame
combined_df.reset_index(drop=True, inplace=True)

# Save the combined DataFrame to a new CSV file
combined_df.to_csv(r'C:\Users\Admin\Downloads\SPARTA\SP702 CAPSTONE\Reviews\restaurant_Reviews.csv', index=False)
```
